```python
import random

random_values = [random.choice([-7, 14]) for _ in range(10)]

total_sum = 0

for day, value in enumerate(random_values, start=1):
    print(f'DAY {day}: {round(value * 5, 2)}')
    total_sum += value

print(f"Sum of all values: {round(total_sum * 5, 2)}")


```

```python
import random

def simulate_trade(prob_win):
    if random.random() < prob_win:
        return 2  # Reward
    else:
        return -1  # Risk

def main():
    prob_win = float(input("Enter your win probability (0 to 1): "))

    total_profit = 0
    risk_reward_ratio = 2

    for day in range(1, 31):
        trade_result = simulate_trade(prob_win)
        total_profit += trade_result

        print(f"Day {day}: Trade result: {trade_result}, Total Profit: {total_profit*1000}")

    final_profit = total_profit * risk_reward_ratio
    print(f"Final Profit (1:2 risk-reward ratio): {final_profit*1000}")

if __name__ == "__main__":
    main()
```
