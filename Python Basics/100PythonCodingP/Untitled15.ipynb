```python
sample = [10, 2, 4, 7, 9, 3, 9, 8, 6, 7,2,3,4,4]
```

```python
import statistics as st

def describe(sample):
    return st.mean(sample), st.median(sample), st.mode(sample)
```

```python
mean, median, mode = describe(sample)
```
