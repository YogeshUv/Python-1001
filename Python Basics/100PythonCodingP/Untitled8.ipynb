```python
nums=[19, 15, 5, 3, 5, 5, 2]

count5,count19 = 0,0
for i in nums:
  if i ==19:
    count19+=1
  if i ==5:
    count5+=1
if count19 >=2:
  print("True")
elif count5 >=3:
  print("True")
else:
  print("False")
```
