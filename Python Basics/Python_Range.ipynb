```python
varname = range(11)

def generate_next():
    for i in varname:
        yield i

print(generate_next())
```

```python
varname = range(11)

def generate_next():
    for i in varname:
        yield i

# Create a generator object
gen = generate_next()

# Print all the values generated by the generator
for value in gen:
    print(value)
```
