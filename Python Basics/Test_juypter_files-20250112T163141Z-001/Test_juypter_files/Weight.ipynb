```python
import datetime

# Define initial weight and increment
wt = 65.5
increment = 0.1

# Get today's date
current_date = datetime.date.today()

# Iterate for 30 days
for i in range(1, 31):
    # Print the current date and weight
    print(f'Day {i} : {current_date}: {wt} KGs')
    
    # Increment the weight and round it to two decimal places
    wt += increment
    wt = round(wt, 2)
    
    # Calculate the date for the next iteration
    current_date += datetime.timedelta(days=1)
```
