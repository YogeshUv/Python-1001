```python
from datetime import datetime

class RequestIDGenerator:
    def __init__(self):
        self.counter = 0

    def generate_request_id(self):
        timestamp = int(datetime.utcnow().timestamp() * 1000)  # Convert current time to milliseconds
        self.counter += 1
        human_readable_format = datetime.utcfromtimestamp(timestamp / 1000).strftime('%Y%m%d%H%M%S%f')[:-3]
        request_id = int(f"{human_readable_format}{self.counter:03d}")
        return request_id

# Example usage
id_generator = RequestIDGenerator()

# Generate request IDs
for _ in range(5):
    request_id = id_generator.generate_request_id()
    print("Request ID:", request_id)
# Request ID: 20231113092708168001 len('20231113092621937001') = 20
```
