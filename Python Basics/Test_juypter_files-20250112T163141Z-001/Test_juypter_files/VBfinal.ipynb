```python
import pandas as pd

def clone_rows_with_column_change(input_df, column_to_change, new_values):
    """
    Clone rows in a DataFrame with one column changed based on a list of values.

    Parameters:
    - input_df (pd.DataFrame): The input DataFrame.
    - column_to_change (str): The name of the column to be changed.
    - new_values (list): The list of new values for the specified column.

    Returns:
    - pd.DataFrame: The resulting DataFrame with cloned rows and the specified column changed.
    """
    new_rows = pd.DataFrame()
    for index, row in input_df.iterrows():
        for new_value in new_values:
            new_row = row.copy()
            new_row[column_to_change] = new_value
            new_rows = new_rows.append(new_row, ignore_index=True)

    return new_rows

# Example usage:
# data = {'Column1': [1, 2, 3],
#         'Column2': ['A', 'B', 'C']}
# df = pd.DataFrame(data)

# new_values = ['X', 'Y', 'Z']
result_df = clone_rows_with_column_change(combined_df_new.iloc[:2], 'bic_vendor_facility', facility_codes)

# Display the resulting DataFrame
# print(result_df)
result_df
```
