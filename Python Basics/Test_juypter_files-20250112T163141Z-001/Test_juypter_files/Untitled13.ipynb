```python
SELECT *
FROM 
    hz_parties,
    hz_party_sites,
    hz_cust_accounts
WHERE 
    hz_parties.party_id = hz_party_sites.party_id
    AND hz_parties.party_id = hz_cust_accounts.party_id;
```

```python
%%time
import pyarrow.parquet as pq
import pandas as pd

# Read the Parquet files into Arrow tables
hz_parties = pq.read_table('hz_parties.parquet')
hz_party_sites = pq.read_table('hz_party_sites.parquet')
hz_cust_accounts = pq.read_table('hz_cust_accounts.parquet')

# # Convert the Arrow tables to Pandas DataFrames
df1 = hz_parties.to_pandas()
df2 = hz_party_sites.to_pandas()
df3 = hz_cust_accounts.to_pandas()

# # Merge the two DataFrames, if needed
merged_df = pd.concat([df1, df2, df3], axis=0)
print(merged_df)
# # Now you have a single DataFrame that contains the data from both Parquet files
```
