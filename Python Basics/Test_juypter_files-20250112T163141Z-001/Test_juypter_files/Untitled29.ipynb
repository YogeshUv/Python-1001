```python
import xml.etree.ElementTree as ET
from io import StringIO
import re

# xml_file_path = r'D:\W\RELEX\RELEX\Future_List_Cost\FUTRLIST.IN07.20231206100039 (Oracle File)'
xml_file_path = 'D:\\W\\RELEX\\RELEX\\Broker and Vendor\\Inbound Files\\BROKER.xml'

# Open the XML file
with open(xml_file_path, 'r', encoding='utf-8') as file:
    xml_data = file.read()

    # Split the XML data into lines
    lines = xml_data.splitlines()
    # Parse the XML data
    # Wrap the XML data in a root element to make it well-formed XML
#     wrapped_xml = f"<root>{xml_data}</root>"
#     # Parse the XML data using ET.parse and StringIO
#     root = ET.parse(StringIO(wrapped_xml)).getroot()
#     root = ET.fromstring(xml_data)
    # Process each line
    for line in lines:
        # Skip empty lines or lines containing only whitespace
        if not line.strip():
            continue

        # Process the line (you can print it here)
        print(line)
        # Define a dictionary to store values
        element_values = {}

        # Regular expression pattern to match the elements
        pattern = re.compile(r"<([^>]+)>([^<]+)</\1>")

        # Find all matches in the text data
        matches = re.findall(pattern, line)

        # Store the matches in the dictionary
        for tag, value in matches:
            element_values[tag] = value.strip()

        # Print the values
        for tag, value in element_values.items():
            print(f"{tag}: {value}")
        print(element_values)
        print('#############################')



```
